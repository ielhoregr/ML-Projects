```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to pandas dataframe 
sonar_data = pd.read_csv('/content/sonar data.csv', header=None)
```

```python
# seperate our featurs and targets 
X = sonar_data.drop(columns = 60, axis=1) #drop the last columns which is our target (true label y)
y = sonar_data[60] 
```

```python
# Training data and test data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.1, stratify= y, random_state=1)
```

test_size = .1 --> take 10 percent of the data as a test


stratify = --> 
This stratify parameter makes a split so that the proportion of values in the sample produced will be the same as the proportion of values provided to parameter stratify.

For example, if variable y is a binary categorical variable with values 0 and 1 and there are 25% of zeros and 75% of ones, stratify=y will make sure that your random split has 25% of 0's and 75% of 1's.

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# training our model 
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# test our model on the training data 

X_train_predection = model.predict(X_train)
X_train_accurecy = accuracy_score(X_train_predection, y_train)
```

```python
print(f'the acuuracy of our model on X_train data is : {round(X_train_accurecy * 100, 2)}%')
```

```python
X_test_predection = model.predict(X_test)
X_test_accuracy = accuracy_score(X_test_predection, y_test)
print(f'the acuuracy of our model on X_test data is : {round(X_test_accuracy * 100, 2)}%')
```

```python
# making a predictive system : 
input_data = (input("Enter your mesures")).split(',')
input_data_final = np.asarray(input_data).reshape(1, -1)
answer = model.predict(input_data_final) 

if(answer[0] == "R"):
  print("Don't worry, it's just a Rock")
else:
  print("Alert, this is a Mine!")
```
